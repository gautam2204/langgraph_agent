### Pydantic are foundation models that are used for data validation at runtime.
    It inherits a BaseModel

```python
from pydantic import BaseModel

class Person(BaseModel):
    name:str
    age:int
    city:str

person = Person(name="gautam",age=21,city="dehradun")
person
    
```

# Problem with dataclass is it do not validates at runtime

```python
from dataclasses import dataclass


@dataclass
class Person1:
    name:str
    age:int
    
person1=Person1(name=123,age=21)
person1
```

### Wi pydantic we can have model wit optional with default 

```python
from typing import Optional


class Person(BaseModel):
    name:str
    age:int
    city:str
    salary:Optional[float] = int
    gender:Optional[bool] = True 
    

person = Person(name="gautam",age=21,city="dehradun",salary=20.22,gender=False)
print(person)

person1 = Person(name="gautam",age=21,city="dehradun",salary=20)
print(person1)
```

### Pyantic also coes with field level constraints 

```python


from pydantic import Field
class Student(BaseModel):
    name:str=Field(description="Name of student")
    rollNo:int=Field(description="Roll Number of student",gt=10,lt=100)

st1 = Student(name="Gauta",rollNo=11)
st1
```
