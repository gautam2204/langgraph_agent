```python
from langchain_community.tools.arxiv.tool import ArxivQueryRun
from langchain_community.utilities import ArxivAPIWrapper

# 1. Setup the wrapper with your config
arxiv_wrapper = ArxivAPIWrapper(
    load_max_docs=2, 
    get_full_documents=True
)

# 2. Create the tool
arxiv_tool = ArxivQueryRun(api_wrapper=arxiv_wrapper)

```

```python
from dotenv import load_dotenv
load_dotenv()
import os

from langchain_tavily import TavilySearch

travily_tool = TavilySearch(
    max_results=5,
    topic="general",
    # include_answer=False,
    # include_raw_content=False,
    # include_images=False,
    # include_image_descriptions=False,
    # search_depth="basic",
    # time_range="day",
    # include_domains=None,
    # exclude_domains=None
)
# travily_tool.invoke({"query": "What happened at the last wimbledon"})
```

```python
# Create list of tools 

tools = [arxiv_tool,travily_tool]
```

```python
# initialize the llm
from langchain_groq import ChatGroq
llm = ChatGroq(model='llama-3.1-8b-instant')
# llm.invoke('where was last wimbeldon held?')
```

```python
llm_with_tools = llm.bind_tools(tools=tools)
llm_with_tools
```

```python
llm_with_tools.invoke('When will next wimbeldon date?')
```
